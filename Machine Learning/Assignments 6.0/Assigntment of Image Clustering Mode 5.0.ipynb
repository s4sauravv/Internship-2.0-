```python
#Lets import Necessory Libraries
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Load the raw image
face = misc.face(gray=True)
```

```python
# Reshape the image into a 2D array of pixels
rows = face.shape[0]
cols = face.shape[1]
image = face.reshape(rows*cols, 1)
```

```python
# Compress the image into 5 clusters using K-Means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(image)
compressed_image = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = compressed_image.reshape(rows, cols)
```

```python
# Visualize the raw and compressed image side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(face, cmap='gray')
axes[0].set_title('Raw Image')
axes[1].imshow(compressed_image, cmap='gray')
axes[1].set_title('Compressed Image (5 Clusters)')
plt.show()
```
